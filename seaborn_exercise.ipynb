## Seaborn Assignment

```python
# import pandas
import pandas as pd

# import seaborn
import seaborn as sns

# import matplotlib
import matplotlib.pyplot as plt

# Apply the default theme
sns.set_theme()

# set size of axes-level plots
sns.set(rc = {'figure.figsize':(15,8)})
```

```python
# import data from github gist csv and drop # column
df = pd.read_csv('https://gist.githubusercontent.com/armgilles/194bcff35001e7eb53a2a8b441e8b2c6/raw/92200bc0a673d5ce2110aaad4544ed6c4010f687/pokemon.csv',sep=',')
df.drop('#',axis=1,inplace=True)
```

### Task 1: Plot a regression line between attack and defense.

```python
# plot scatterplot and regression - this is an axes-level function so we can set the title inline
sns.regplot(data=df, x='Attack', y='Defense').set_title('Attack vs. Defense')
```

### Task 2: Plot a regression line between attack and defense while displaying a distinction between the different generations. 

```python
# plot scatterplot and regressions by generation
plot = sns.lmplot(data=df, x='Attack', y='Defense', hue='Generation', height=8, aspect=15/8)

# as this is a figure-level function, set title seprately
plot.fig.suptitle('Attack vs Defense by Generation')
```

### Task 3: Make a boxplot for each feature of dataframe (without Generation and Legendary features) in one plot.

```python
# filter out Generation and Legendary features
df_box = df.iloc[:,0:10]

# plot our new dataframe with title inline
sns.boxplot(data=df_box).set_title('Feature Overview: Boxplot')
```

### Task 4: Plot a correlation heatmap of the pokemon dataframe (not including the features Generation and Legendary).

```python
# create correlation dataframe and save it
df_corr = df_box.corr()
df_corr
```

```python
# create heatmap from correlation dataframe
heatmap = sns.heatmap(df_corr, vmin=-1, vmax=1, annot=True)

# as we've saved our heatmap to a variable, let's demonstrate another way to set the title on an axes-level function
heatmap.set_title('Correlation Heatmap', fontdict={'fontsize':18}, pad=12);
```
