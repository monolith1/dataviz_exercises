## Matplotlib Assignment

```python
# import pandas
import pandas as pd

# import matplotlib
import matplotlib.pyplot as plt
```

```python
# set figsize
plt.rcParams["figure.figsize"] = (15,8)
```

```python
# import data - available at 
df=pd.read_csv('company_sales_data.csv')
```

```python
# preview first 5 entries of dataframe
df.head()
```

```python
# check dtypes
df.dtypes
```

```python
# check for null entries
df.isnull().sum()
```

### Task 1: Read the `total_profit` of all months and display it using a line plot.

```python
# extract data to variables
x = df.month_number
y = df.total_profit

# create plot
plt.plot(x,y,'red')

# set title and x/y labels
plt.title('Total Profit by Month')
plt.xlabel('Month')
plt.ylabel('Profit')

# show the plot
plt.show()
```

### Task 2: Read all the different product sales data and display it using a multiline plot.

```python
# extract data to variables
y1 = df.facecream
y2 = df.facewash
y3 = df.toothpaste
y4 = df.bathingsoap
y5 = df.shampoo
y6 = df.moisturizer

# plot each variable with title
plt.plot(x,y1, 'blue', label='facecream')
plt.plot(x,y2, 'olive', label='facewash')
plt.plot(x,y3, 'cyan', label='toothpaste')
plt.plot(x,y4, 'pink', label='bathingsoap')
plt.plot(x,y5, 'purple', label='shampoo')
plt.plot(x,y6, 'green', label='moisturizer')

# show the plot labels
plt.legend()

# set title and x/y labels
plt.title('Unit Sales by Month')
plt.xlabel('Month')
plt.ylabel('Unit Sales')

# show the plot
plt.show()
```

### Task 3: Read the `total_profit` of each month and display it using a histogram to find out which profit ranges are the most common.

```python
# extract data to variable
profit = df.total_profit

# set bins explicitly
bins = [100000, 150000, 200000, 250000, 300000, 350000]

# plot histogram
plt.hist(profit, bins)

# set title and x/y labels
plt.title('Total Profit by Month')
plt.xlabel('Month')
plt.ylabel('Total Profit')

# show the plot
plt.show()
```
